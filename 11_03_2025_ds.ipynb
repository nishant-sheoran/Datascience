Data Cleaning

```python
import pandas as pd
```

```python
data = {'Name': ['Alice', 'Bob', None, 'David'],
        'Age': [25, None, 30, 40],
        'City': ['New York', 'Paris', 'London', None]}
df = pd.DataFrame(data)
print(df)
```

```python
df_cleaned = df.dropna() # NA rows are dropped
print(df_cleaned)
```

```python
df_filled = df.fillna({'Name': 'Unknown', 'Age': df['Age'].mean(), 'City': 'Unknown'})
print(df_filled)
```

Data Transformation

```python
#Applying Functions to Columns (apply(), map())
df_filled['Age_Category'] = df_filled['Age'].apply(lambda x:'Young' if x < 30 else 'Old')
df_filled
```

```python
#Changing Data Types (astype())
df_filled['Age'] = df_filled['Age'].astype(int) # Convert float to integer
df_filled
```

```python
#Renaming Columns (rename())
df_filled = df_filled.rename(columns={'Name': 'Full_Name'})
df_filled
```

```python
#Creating New Columns (assign())
df_filled = df_filled.assign(Age_Squared=df['Age'] ** 2)
df_filled
```

Merging Data

```python
# Using merge() for SQL-like Joins
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name':
['Alice', 'Bob', 'Charlie']})
print(df1)
print("\n")
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Score':
[85, 90, 75]})
print(df2)
print("\n")

merged_df = pd.merge(df1, df2, on='ID', how='left') # Left join
print(merged_df)
print("\n")
merged_df1 = pd.merge(df1, df2, on='ID', how='right') # Right join
print(merged_df1)
print("\n")
merged_df2 = pd.merge(df1, df2, on='ID', how='inner') # Inner join
print(merged_df2)
print("\n")
merged_df3 = pd.merge(df1, df2, on='ID', how='outer') # Outer join
print(merged_df3)
```

Row wise concatenation

```python
df1 = pd.DataFrame({'ID': [1, 2], 'Name':
['Alice', 'Bob']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name':
['Charlie', 'David']})

concat_df = pd.concat([df1, df2], ignore_index = True) # Stacks them vertically
print(concat_df)
```
